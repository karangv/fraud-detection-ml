# Fraud Detection Decision Dashboard

## Project Overview
Fraud detection is a classic machine learning problem where the objective is to identify rare fraudulent transactions among a large volume of legitimate activity.

This project focuses not only on predicting fraud, but on building a **decision-support system** that helps stakeholders understand:
- Model confidence
- Tradeoffs between false positives and false negatives
- How decision thresholds impact business outcomes

## Problem Statement

Given a dataset of historical transactions, the goal is to build a machine learning model that predicts whether a transaction is fraudulent.

This is a **binary classification problem** with the following challenges:
- Extreme class imbalance (fraud is rare)
- High cost of false negatives (missed fraud)
- Non-trivial cost of false positives (customer friction)

## Evaluation Considerations

Accuracy is not an appropriate metric for this problem due to severe class imbalance.

Instead, we focus on:
- Precision
- Recall
- F1-score
- Precision-Recall AUC
- Cost-sensitive evaluation

The choice of decision threshold is treated as a business decision rather than a purely statistical one.

## Constraints and Assumptions

- Fraud rate is assumed to be < 1%
- Predictions should be explainable
- Model latency should support near real-time decisions
- Dataset is assumed to be static and labeled
